## Word count of Shakespeare's corpus

```python
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark import SparkContext
import pandas as pd
```

```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

### Load the corpus

```python
txt = spark.read.text("../../assets/datasets/pg100.txt", wholetext=True)
```

```python
corpusWords = txt.select(explode(split(txt.value, r"\s+")).alias("word"))
```

```python
corpusWords = corpusWords.withColumn("letter", lower(substring(corpusWords["word"], 1, 1)))
```

### Group corpus words by the letter they start and get the most popular words and starting letters

```python
words = corpusWords.groupBy("word").count().sort(desc("count"))
words.show(10)
```

```python
letters = corpusWords.groupby("letter").count().sort(desc("count"))
letters.show(10)
```
