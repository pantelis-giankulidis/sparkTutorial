## Page Rank

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
G = nx.read_edgelist("../../assets/datasets/web-Stanford.txt", create_using=nx.DiGraph())

def print_graph_info(G, directed=True):
  print("Number of nodes:", len(G.nodes))
  print("Number of edges:", len(G.edges))
  if directed:
    print("Average in-degree:", sum(dict(G.in_degree).values()) / len(G.nodes))
    print("Average out-degree:", sum(dict(G.out_degree).values()) / len(G.nodes))
  else:
    print("Average degree:", sum(dict(G.degree).values()) / len(G.nodes))
```

```python
print_graph_info(G)
```

We have to extract the largest weakly connected component from the graph.

```python
largest_cc = max(nx.weakly_connected_components(G), key=len)
G = G.subgraph(largest_cc)
print_graph_info(G, directed=True)
```

Get the page rank of the graph, using the available networkx function.

```python
pr = nx.pagerank(G, alpha=0.9)
```

### Generating random and barabasi-albert graphs
In this section we will generate random and barabasi-albert graphs and compute the PageRank for each of them.

```python
random_G = nx.fast_gnp_random_graph(len(G.nodes), 0.00008)
print_graph_info(random_G, directed=False)

barabasi_albert_G = nx.barabasi_albert_graph(len(G.nodes), 1) # choose m such that the number of edges is close to the number of edges in G 
print_graph_info(barabasi_albert_G, directed=False)
```

Computing the page ranks for both graphs

```python
random_G_pr = nx.pagerank(random_G, alpha = 0.9)
barabasi_albert_G_pr = nx.pagerank(barabasi_albert_G, alpha = 0.9)
```

Compare the three pageRank vectors, from the three different graphs, actual, random and barabasi-albert, using the cosine similarity metric.

```python
pageRank = np.array(list(pr.values())).reshape(-1, 1)
pageRank_random = np.array(list(random_G_pr.values())).reshape(-1, 1)
pageRank_barabasi_albert = np.array(list(barabasi_albert_G_pr.values())).reshape(-1, 1)

pageRank = -np.sort(-pageRank, axis=0)
pageRank_random = -np.sort(-pageRank_random, axis=0)
pageRank_barabasi_albert = -np.sort(-pageRank_barabasi_albert, axis=0)


p = np.linalg.norm(pageRank)
r = np.linalg.norm(pageRank_random)
ba = np.linalg.norm(pageRank_barabasi_albert)

cosine_sim_pageRank_random = np.dot(pageRank.flatten(), pageRank_random.flatten()) / (p * r)
cosine_sim_pageRank_barabasi_albert = np.dot(pageRank.flatten(), pageRank_barabasi_albert.flatten()) / (p * ba)
cosine_sim_random_barabasi_albert = np.dot(pageRank_random.flatten(), pageRank_barabasi_albert.flatten()) / (r * ba)

print("Cosine similarity between PageRank and Random Graph:", cosine_sim_pageRank_random)
print("Cosine similarity between PageRank and Barabasi-Albert Graph:", cosine_sim_pageRank_barabasi_albert)
print("Cosine similarity between Random Graph and Barabasi-Albert Graph:", cosine_sim_random_barabasi_albert)

```
